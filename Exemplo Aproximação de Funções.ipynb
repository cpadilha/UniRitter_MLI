```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0, 20, 1e-3)
```

```python
y = np.sin(x)
```

```python
x = x.reshape((10000, 1))
```

```python
plt.plot(x, y)
```

<h2>Criando o modelo</h2>

```python
from sklearn.neural_network import MLPRegressor
```

```python
mlp = MLPRegressor()
mlp.fit(x, y)
```

<h2>Testando</h2>

```python
x2 = np.arange(11, 20, 5e-2)
x2 = x2.reshape((len(x2), 1))
```

```python
y2 = mlp.predict(x2)
```

<h2>Real vs Predito</h2>

```python
plt.clf()
plt.plot(x, y)
plt.scatter(x2, y2, marker='o', c='r')
plt.show()
```

<h2>Previs√£o para valores fora da faixa treinada</h2>

```python
x = np.arange(0, 10, 1e-3)
x = x.reshape((10000, 1))

y = np.sin(x)
```

```python
trans_x = np.hstack([np.cos(x), np.sin(x)])
```

```python
mlp = MLPRegressor()
mlp.fit(trans_x, y)
```

```python
x2 = np.arange(11, 20, 5e-2)
x2 = x2.reshape((len(x2), 1))
trans_x2 = np.hstack([np.cos(x2), np.sin(x2)])
```

```python
y2 = mlp.predict(trans_x2)
```

```python
x = np.arange(0, 20, 1e-3)
y = np.sin(np.cos(x))

plt.plot(x, y)
plt.plot(x2, y2, c='r')
plt.show()
```
