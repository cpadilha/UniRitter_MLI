```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt("data_1024.csv.txt", delimiter='\t', skip_header=0)
```

```python
data = data[1:, 1:]
```

```python
plt.scatter(data[:,0], data[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(min_samples=15, eps=10.0)
```

```python
dbscan.fit(data)
```

```python
labels = dbscan.labels_
```

```python
np.unique(labels)
```

```python
np.count_nonzero(labels == -1)
```

```python
np.count_nonzero(labels != -1)
```

```python
data_0 = data[dbscan.labels_ == 0]
data_1 = data[dbscan.labels_ == 1]
data_2 = data[dbscan.labels_ == 2]
```

```python
plt.scatter(data[:,0], data[:,1])
plt.scatter(data_0[:,0], data_0[:,1])
plt.scatter(data_1[:,0], data_1[:,1])
plt.scatter(data_2[:,0], data_2[:,1])
```
